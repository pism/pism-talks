```python
import numpy as np
import pylab as plt
import scipy.stats as stats
```

```python
fontsize = 16
params = {
    "axes.labelsize": fontsize,
    "font.size": fontsize,
    "legend.fontsize": fontsize,
    "font.size": fontsize,
}

plt.rcParams.update(params)
```

```python
x = np.linspace(-12, 12, 1001)
fig = plt.figure()
ax = fig.add_subplot(111)
plt.xticks([]),plt.yticks([])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Contribution to sea-level (cm SLE)")
ax.set_ylabel("Probability density")

pdf = stats.norm.pdf(x, -1, 1.5)
ax.plot(x, pdf, label="ISMIP6")
ax.axvline(-1, c="0.5", ls="dotted", lw=0.5)
legend = ax.legend(loc="upper left")
legend.get_frame().set_linewidth(0.0)
legend.get_frame().set_alpha(0.0)
fig.savefig("figures/sle_pdf_initial.pdf")

pdf = stats.norm.pdf(x, 0, 4)
ax.plot(x, pdf, label="ISMIP7")
legend = ax.legend(loc="upper left")

legend.get_frame().set_linewidth(0.0)
legend.get_frame().set_alpha(0.0)
fig.savefig("figures/sle_pdf_uq.pdf")

pdf = stats.norm.pdf(x, 2, 1)
ax.axvline(2, c="0.5", ls="dotted", lw=0.5)
ax.plot(x, pdf, label="ISMIP7+Calib.")
legend = ax.legend(loc="upper left")

legend.get_frame().set_linewidth(0.0)
legend.get_frame().set_alpha(0.0)

fig.savefig("figures/sle_pdf_calib.pdf")


```
