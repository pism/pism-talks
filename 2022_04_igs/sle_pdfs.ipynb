```python
import numpy as np
import pylab as plt
import scipy.stats as stats
```

```python
fontsize = 18
params = {
    "axes.labelsize": fontsize,
    "font.size": fontsize,
    "legend.fontsize": fontsize,
    "font.size": fontsize,
}

plt.rcParams.update(params)
```

```python
x = np.linspace(-10, 10, 1001)
fig = plt.figure()
ax = fig.add_subplot(111)
plt.xticks([]),plt.yticks([])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Contribution to sea-level (cm SLE)")
ax.set_ylabel("Probability density")

pdf = stats.norm.pdf(x, 1, 2)
ax.plot(x, pdf, label="Initial")
legend = ax.legend()
legend.get_frame().set_linewidth(0.0)
legend.get_frame().set_alpha(0.0)
fig.savefig("figures/sle_pdf_initial.pdf")

pdf = stats.norm.pdf(x, 1, 4)
ax.plot(x, pdf, label="UQ")
legend = ax.legend()
legend.get_frame().set_linewidth(0.0)
legend.get_frame().set_alpha(0.0)
fig.savefig("figures/sle_pdf_uq.pdf")

pdf = stats.norm.pdf(x, 1, 1)
ax.plot(x, pdf, label="Calibrated")
legend = ax.legend()
legend.get_frame().set_linewidth(0.0)
legend.get_frame().set_alpha(0.0)

fig.savefig("figures/sle_pdf_calib.pdf")


```
